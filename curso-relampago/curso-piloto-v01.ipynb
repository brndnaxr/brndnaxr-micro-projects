## Objetivos do curso <a class="anchor" id="objetivos"></a>

* Aprender o ``básico de programação`` **para estatística**.
* Aprender os **fundamentos** da estatística e probabilidade. De forma sólida.
* Conseguir resolver **problemas reais** utilizando ``python`` e ``estatística``.
* Com poucas linhas de código e **muito fundamento**, **gerar resultado** no seu trabalho, ou no seu dia a dia.
* Compreender como áreas distintas se relacionam: **estatística** e **programação**.
