<h2 id="escape">Escape Sequences</h2>


As barras invertidas representam o início das sequências de escape. As sequências de escape representam strings que podem ser difíceis de inserir. Por exemplo, a barra invertida "n" representa uma nova linha. A saída é dada por uma nova linha após a barra invertida "n" ser encontrada:

```python
# New line escape sequence

print(" Michael Jackson \n is the best" )
```

Da mesma forma, a barra invertida "t" representa uma guia:

```python
# Tab escape sequence

print(" Michael Jackson \t is the best" )
```

Se você quiser colocar uma barra invertida em sua string, use uma barra invertida dupla:

```python
# Include back slash in string

print(" Michael Jackson \\ is the best" )
```

Também podemos colocar um "r" antes da string para exibir a barra invertida:

```python
# r will tell python that string will be display as raw string

print(r" Michael Jackson \ is the best" )
```

<h2 id="operations">String Operations</h2>


Existem muitos métodos de operação de string em Python que podem ser usados ​​para manipular os dados. Vamos usar algumas operações básicas de string nos dados.

Vamos tentar com o método <code> superior </code>; este método converte caracteres minúsculos em caracteres maiúsculos:

```python
# Convert all the characters in string to upper case

a = "Thriller is the sixth studio album"
print("before upper:", a)
b = a.upper()
print("After upper:", b)
```

O método <code> replace </code> substitui um segmento da string, ou seja, uma substring por uma nova string. Introduzimos a parte da string que gostaríamos de alterar. O segundo argumento é com o que gostaríamos de trocar o segmento, e o resultado é uma nova string com o segmento alterado:

```python
# Replace the old substring with the new target substring is the segment has been found in the string

a = "Michael Jackson is the best"
b = a.replace('Michael', 'Janet')
b
```

O método <code> find </code> encontra uma sub-string. O argumento é a substring que você gostaria de encontrar e a saída é o primeiro índice da sequência. Podemos encontrar a sub-string <code> jack </code> ou <code> el <code>.

<img src="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0101EN-SkillsNetwork/labs/Module%201/images/StringsFind.png" width="600" align="center" />


```python
# Find the substring in the string. Only the index of the first elment of substring in string will be the output

name = "Michael Jackson"
name.find('el')
```

```python
# Find the substring in the string.

name.find('Jack')
```

Se a sequência secundária não estiver na sequência, a saída será negativa. Por exemplo, a string 'Jasdfasdasdf' não é uma substring:

```python
# If cannot find the substring in the string

name.find('Jasdfasdasdf')
```

```python
# imagine uma lista com o salário dos funcionários

salarios = [1290, 3250, 2890, 4130, 1540]
```

De repente, eles recebem um aumento de 25%. E, agora, precisamos saber os valores dos salários atualizados.

```python
# 1 método - map
```

```python
def vezes_aumento (salario_velho):
    '''
    salario novo = salario velho * 1.25
    '''
    salario_novo = salario_velho + (salario_velho * 0.25)
    return salario_novo
```

```python
salarios_atual = map(vezes_aumento, salarios)
```

```python
print(list(salarios_atual))
```

```python
# 2º método - lambda
```

```python
salarios_atual_v2 = map(lambda x: x + (x * 0.25), salarios)
```

```python
print(list(salarios_atual_v2))
```
