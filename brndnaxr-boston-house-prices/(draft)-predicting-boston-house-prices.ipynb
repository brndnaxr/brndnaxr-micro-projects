Conhecendo as colunas da base de dados

**`CRIM`**: Taxa de criminalidade per capita por regiao.

**`ZN`**: Proporção de terrenos residenciais divididos por lotes com mais de 25.000 pés quadrados.

**`INDUS`**: Essa é a proporção de hectares de negócios não comerciais por regiao.

**`CHAS`**: variável fictícia Charles River (= 1 se o trecho limita o rio; 0 caso contrário)

**`NOX`**: concentração de óxido nítrico (partes por 10 milhões)

**`RM`**: Número médio de quartos entre as casas do bairro

**`Age`**: proporção de unidades ocupadas pelos proprietários construídas antes de 1940

**`DIS`**: distâncias ponderadas para cinco centros de emprego em Boston

**`RAD`**: Índice de acessibilidade às rodovias radiais

**`IMPOSTO`**: taxa do imposto sobre a propriedade de valor total por US $ 10.000

**`B`**: 1000 (Bk - 0,63) ², onde Bk é a proporção de pessoas de descendência afro-americana por regiao

**`PTRATIO`**: Bairros com maior proporção de alunos para professores (maior valor de 'PTRATIO')

**`LSTAT`**: porcentagem de status mais baixo da população

**`MEDV`**: valor médio de casas ocupadas pelos proprietários em US $ 1000

**Definindo um baseline**
* Uma baseline é importante para ter marcos no projeto.
* Permite uma explicação fácil para todos os envolvidos.
* É algo que sempre tentaremos ganhar na medida do possível.

**Sobre a Baseline**
* A nossa *baseline* será definida como o valor do imóvel diretamente proporcional ao número de quartos.
* Portanto, nossa solução deve bater essa.
* Porque caso contrário, não faz sentido utilizar metodologias mais complexas para resolver um problema que é simples.

```python
# converte os dados
data.RM = data.RM.astype(int)
```

```python
data.info()  # visualizando o resultado
```

**Vamos definir categorias**
* Imóveis com até 4 quartos = pequenos
* Imóveis de 4 até 7 quartos = médios
* Imóveis acima de 7 quartos = grandes

```python
# definindo a regra para categorizar os dados
categorias = [ ]
```

```python
# fazendo append() na lista categorias
for i in data.RM.iteritems():
  valor = (i[1])
  if valor <= 4:
    categorias.append('Pequeno')
  elif valor < 7:
    categorias.append('Medio')
  else:
    categorias.append('Grande')
```

```python
categorias   # visualizando a lista
```

```python
# criando a coluna 'categorias' com a lista criada
data['Categorias'] = categorias
data.head()
```

```python
data.Categorias.value_counts()   # contando quantos têm de cada
```

Agrupando por categoria e calculando a média de cada categoria

```python
medias_categoria = data.groupby(by = 'Categorias')['MEDV'].mean()
```

```python
medias_categoria   # viz
```

```python
# criando o dicionario com chaves 'medio', 'grande' e 'pequeno' - seus valores
dic_baseline = {'Grande': medias_categoria[0], 'Medio': medias_categoria[1], 'Pequeno': medias_categoria[2]}
dic_baseline  # viz
```

```python
# vamos criar uma função que consulta o dicionário e retorna a baseline
def retorna_baseline(num_quartos):
  '''
  Traz o preço médio do imóvel de acordo com a sua categoria
  '''
  if num_quartos <= 4:
    return dic_baseline.get('Pequeno')
  elif num_quartos < 7:
    return dic_baseline.get('Medio')
  else:
    return dic_baseline.get('Grande')
```

```python
# chama a função baseline
retorna_baseline(3)
```

# Aula 3

## Trabalhando com a baseline

### Preparando os dados

```python
# variável 'y' é a nossa var. alvo
y = data['MEDV']
```

```python
# remove as colunas colineares variáveis alvo e a coluna categorias
X = data.drop(['RAD','TAX','MEDV','DIS','AGE','ZN','Categorias'], axis = 1)
X.head()
```

### Separando os conjuntos de treino e teste

```python
# carrega o método train_test_split
from sklearn.model_selection import train_test_split
```

```python
# divide os dados entre o conjunto de treino e teste, 80% e 20% respectivamente
# define qualquer valor para o parâmetro random_state
# random_state é o tal do seed() -> para experimentos com 'mesma aletoriedade'
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 5)
```

```python
# imprime a quantidade de linhas de cada conjunto que foi separado
print('X_train -> número de linhas e colunas = {}' .format(X_train.shape))
print('X_test -> número de linhas e colunas = {}' .format(X_test.shape))
print('y_train -> número de linhas e colunas = {}' .format(y_train.shape))
print('y_test -> número de linhas e colunas = {}' .format(y_test.shape))
```

### Modelo de baseline

```python
X_test.head()   # viz
```

```python
# definindo uma lista de predições
p = []
```

```python
# para cada elemento do teste alimenta a lista 'p' de predições
for i in X_test.RM.iteritems():
  n_quartos = i[1]
  p.append(retorna_baseline(n_quartos))
```

```python
p[:10]   # viz
```

### Atribuindo resultados a um dataframe
* Manteremos um log ao longo do projeto
* Permite a evolução do projeto estipulado um baseline

```python
# cria um dataframe vazio
df_results = pd.DataFrame()

# adiciona a coluna valor real
df_results['valor_real'] = y_test.values

# cria a coluna valor_predito_baseline com as predições
df_results['valor_predito_baseline'] = p

# viz
df_results.head(10)
```

### Plotar o valor real e o baseline

```python
import plotly.graph_objects as go

# cria 'traces'
fig = go.Figure()

# linha com os dados de teste
fig.add_trace(go.Scatter(x = df_results.index,
                         y = df_results.valor_real,
                         mode = 'lines + markers',
                         name = 'Valor Real'))

# linha com os dados preditos
fig.add_trace(go.Scatter(x = df_results.index,
                         y = df_results.valor_predito_baseline,
                         mode = 'lines + markers',
                         name = 'Valor Predito Baseline'))

# viz
fig.show()
```

### Avaliação do modelo baseline

```python
# calcula a métrica RMSE
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
# criando objeto RMSE
RMSE = (np.sqrt(mean_squared_error(y_test, p)))
```

```python
# viz a performance do modelo
print('Performance do modelo baseline:')
print('\nRMSE é: {}'.format(RMSE))
```

### **Tuning do Algoritmo de Machine Learning**

**Parâmetros do Random Forest**
* O parâmetro ``n_estimators`` define a quantidade de árvores que serão usadas para criação dos modelos. Quanto maior o número de árvores, maior será o número de modelos no ensemble.
* O parâmetro ``max_deph`` define a profundidade da árvore, quanto mais profunda a árovre, mais especializado nos dados de treino o nosso modelo estará.
* O parâmetro ``max_features`` define a quantidade máxima de nós que serão usados para separar os níveis folha da árovre.
